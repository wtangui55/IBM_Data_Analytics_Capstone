## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# SQL query to select the ConvertedComp column from the master table
QUERY = """
SELECT ConvertedComp
FROM master
"""

# Uses pandas to run the SQL query and loads the data into a DataFrame
converted_comp_data = pd.read_sql_query(QUERY, conn)
# Displays the first 5 rows of the DataFrame
converted_comp_data.head(5)
```

```python
# Plots histogram of ConvertedComp data
plt.figure(figsize=(10, 6))
plt.hist(converted_comp_data['ConvertedComp'].dropna(), bins=50, edgecolor='k', color='skyblue')
plt.title('Distribution of Annual Compensation (USD)')
plt.xlabel('Annual Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Sql query to select Age column from the master table
QUERY = """
Select Age
FROM master
"""
# Uses pandas to run the SQL query and load it into a DataFrame
age_data = pd.read_sql_query(QUERY, conn)

# Displays the firt 5 rows of the DataFrame
age_data.head(5)
```

```python
# Plots box plot of Age data
plt.figure(figsize=(10, 6))
plt.boxplot(age_data['Age'].dropna(), vert=True, patch_artist=True)
plt.title('Box Plot of Age Among Survey Respondents')
plt.ylabel('Age')
plt.xticks([1], ['Age'])
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
database_desire=pd.read_sql_query(QUERY, conn)
database_desire.head()
```

```python
# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(database_desire['Count'], labels=database_desire['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# SQL query to find the rank of Python
query_python_rank = """
SELECT LanguageDesireNextYear, COUNT(*) AS Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
"""

# Execute the query and store the result in a DataFrame
python_rank_data = pd.read_sql_query(query_python_rank, conn)

# Find the rank of Python
python_rank = python_rank_data[python_rank_data['LanguageDesireNextYear'] == 'Python'].index[0] + 1

# Display the rank of Python
print(f"Python is ranked {python_rank} among the languages that respondents wish to learn next year.")
```

```python
# SQL query to count how many respondents currently work with SQL
query_sql_users = """
SELECT COUNT(*) AS SQLUsers
FROM LanguageWorkedWith
WHERE LanguageWorkedWith = 'SQL'
"""

# Execute the query and store the result in a DataFrame
sql_users_data = pd.read_sql_query(query_sql_users, conn)

# Display the number of SQL users
sql_users = sql_users_data['SQLUsers'].iloc[0]
print(f"{sql_users} respondents indicated that they currently work with 'SQL'.")
```

```python
# SQL query to find the number of respondents who work with 'MySQL' only
query_mysql_only_users = """
SELECT COUNT(DISTINCT Respondent) AS MySQLOnlyUsers
FROM DatabaseWorkedWith
WHERE DatabaseWorkedWith = 'MySQL'
AND Respondent NOT IN (
    SELECT Respondent FROM DatabaseWorkedWith WHERE DatabaseWorkedWith != 'MySQL'
)
"""

# Execute the query and store the result in a DataFrame
mysql_only_users_data = pd.read_sql_query(query_mysql_only_users, conn)

# Display the number of MySQL only users
mysql_only_users = mysql_only_users_data['MySQLOnlyUsers'].iloc[0]
print(f"{mysql_only_users} respondents indicated that they work with 'MySQL' only.")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# This SQL query extraccts WorkWeekHrs and CodeRevHrs for respondents aged between 30 and 35
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
# Run the SQL query and load the data into a DataFrame
age_group_data = pd.read_sql_query(QUERY, conn)

# Calculate the median values for WorkWeekHrs and CodeRevHrs
medians = age_group_data.median()
```

```python
# Plotting the stacked chart
plt.bar('Age Group 30-35', medians['WorkWeekHrs'], color='skyblue', label='WorkWeekHrs')
plt.bar('Age Group 30-35', medians['CodeRevHrs'], bottom=medians['WorkWeekHrs'], color='lightgreen', label='CodeRevHrs')
plt.ylabel('Hours')
plt.title('Median Work Hours and Code Review Hours for Age 30 to 35')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# SQL query to extract Age and ConvertedComp for respondents aged between 45 and 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Run the SQL query and load the data into a DataFrame
age_compensation_data = pd.read_sql_query(QUERY, conn)

# Group by Age and calculate median ConvertedComp for each age
median_compensation_by_age = age_compensation_data.groupby('Age').median()

# Reset index to use Age as a column for plotting
median_compensation_by_age.reset_index(inplace=True)
```

```python
# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_compensation_by_age['Age'], median_compensation_by_age['ConvertedComp'], marker='o')
plt.title('Median Annual Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Annual Compensation (USD)')
plt.grid(True)
plt.show()
```

```python
# your code goes here
# SQL query to extract Age and ConvertedComp for respondents aged between 45 and 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
"""

# Run the SQL query and load the data into a DataFrame
age_comp_data = pd.read_sql_query(QUERY, conn)

# Group by Age and calculate median ConvertedComp for each age
median_comp_by_age = age_comp_data.groupby('Age').median()

# Reset index to use Age as a column for plotting
median_comp_by_age.reset_index(inplace=True)
```

```python
# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedComp'], marker='o')
plt.title('Median Annual Compensation for Ages 25 to 30')
plt.xlabel('Age')
plt.ylabel('Median Annual Compensation (USD)')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as RespondentCount
FROM master
GROUP BY MainBranch
ORDER BY RespondentCount DESC
"""

main_branch_data = pd.read_sql_query(QUERY, conn)

main_branch_data.head(5)
```

```python
# Plotting the horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(main_branch_data['MainBranch'], main_branch_data['RespondentCount'], color='green')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.title('Number of Respondents by Main Branch')
plt.gca().invert_yaxis()  # Invert y-axis to have the largest group on top
plt.show()
```

```python
QUERY = """
SELECT DevType, COUNT(*) as Count
FROM DevType
WHERE DevType IN ('Developer, mobile', 'Developer, back-end', 'Developer, full-stack', 'Developer, front-end')
GROUP BY DevType
ORDER BY Count DESC
"""
# Execute the query and store the result in a DataFrame
dev_type_data = pd.read_sql_query(QUERY, conn)
```

```python
# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
plt.barh(dev_type_data['DevType'], dev_type_data['Count'], color='skyblue')
plt.xlabel('Number of Respondents')
plt.title('Number of Respondents by Developer Type')
plt.gca().invert_yaxis()  # To display the bar with the highest count on top
plt.show()
```
